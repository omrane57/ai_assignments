# Assignment 3 : (Scatterplot)

```python
# Task: 
# Create a scatter plot to show the relationship between total bill and tip amount.(Note : Use ‘tips’ dataset from seaborn datasets)
#     • Change the color of the data point
#     • Change the marker
#     • Change the size of the data points based on ‘size’ column
#     • Change the opacity of the graph

tips=sns.load_dataset('tips')
tips['sex_numerical'] = np.where(tips['sex'] == 'Male', 0, 1)
scat=plt.scatter(tips['total_bill'],tips['tip'],alpha=0.6,s=tips['size']*20,label=tips['sex_numerical'],cmap='coolwarm',c=tips['sex_numerical'])

cbar = plt.colorbar(scat)
cbar.set_ticks([0, 1])
cbar.set_ticklabels(['Male', 'Female'])
plt.show()
```

```python
iris_dataset=sns.load_dataset('iris')
iris_dataset.head(5)
unique_species=iris_dataset['species'].unique()
iris_dataset['unique_species_numeric_value']=np.where(iris_dataset['species']=='setosa',0,np.where(iris_dataset['species']=='versicolor',1,np.where(iris_dataset['species']=='virginica',2,3)))
```

```python
overs_df=pd.read_csv('overs.csv')
overs_df
```

```python

overs_df.head(5)
overs_df_first_innings=overs_df[overs_df['innings']==1]
total_run_scored_each_over=overs_df_first_innings.groupby('overs')['total_run'].sum()
total_run_scored_each_over
```

```python
plt.figure(figsize=(15,10))
overs = np.arange(1, len(total_run_scored_each_over) + 1) 

data_reshaped = total_run_scored_each_over.to_numpy().reshape(1, -1) 
plt.imshow(data_reshaped)
plt.imshow(data_reshaped, aspect='auto', cmap='viridis')


plt.colorbar(label='Total Runs Scored')


plt.xticks(ticks=np.arange(len(total_run_scored_each_over)), labels=np.arange(1, len(total_run_scored_each_over) + 1))


plt.xlabel('Overs')
plt.title('Total Runs Scored Each Over')

plt.show()
```

```python
overs_df_six=overs_df[(overs_df['batsman_run']==6 )& (overs_df['non_boundary']==1)]
overs_df_six.shape

pivoted_table=overs_df_six.pivot_table(aggfunc='count',index='overs',values="",columns='batter	')
pivoted_table.fillna(0,inplace=True)
heat=plt.imshow(pivoted_table)
plt.colorbar(heat)
plt.show()
```

```python

overs_df_first_innings=overs_df[(overs_df['innings']==1) &(overs_df['batter']=='YBK Jaiswal')]
pivoted_table=overs_df_first_innings.pivot_table(aggfunc='sum',index='ballnumber',values="total_run")
pivoted_table.fillna(0,inplace=True)

heats=plt.imshow(pivoted_table)
plt.colorbar(heats)
plt.show()
```

```python
overs_df_second_innings=overs_df[(overs_df['innings']==2)&(overs_df['extra_type'].notna())]
pivoted_tables=overs_df_second_innings.pivot_table(aggfunc='count',index='overs',columns="extra_type")
plt.imshow(pivoted_tables)
```

```python
overs_df_innings_both=overs_df[((overs_df['innings']==1)|(overs_df['innings']==2))&(overs_df['isWicketDelivery']==1)]
pivoted_tables=overs_df_innings_both.pivot_table(aggfunc='count',index='overs',columns="innings",values="isWicketDelivery")
pivoted_tables
plt.imshow(pivoted_tables)
```
